```python
# NIM/Nama  : 16521432/Syafiq Ziyadul Arifin
# Tanggal   : 1 Desember 2021
# Deskripsi : Diberikan data pengguna kartu kredit suatu bank
# Judul     : Problem 3 Praktikum 5 Praktikum Pengenalan Komputasi Pengantar 2021-2022

# Kamus
# import pandas as pd: mengimpor library pandas dan meng-assign pandas sebagai pd untuk digunakan
# df: variabel yang membaca data "P05-01.csv"
# .read_csv(): membaca data
# .loc(): menampilkan data tertentu
# .plot(): menampilkan grafik/diagram tertentu
# kind: jenis grafik
# bins: range suatu bar pada histogram
# rwidth: jarak tiap bar pada histogram
# x: kolom data yang digunakan pada sumbu x
# y: kolom data yang digunakan pada sumbu y
# stacked: menentukan apakah y pada grafik akan membentuk stack (bertumpuk) atau tidak
# .value_counts(): membaca banyaknya data sesuai dengan nilai tiap data
# .groupby(): mengelompokkan data sesuai kolom tertentu
# .size(): menentukan jumlah data
# .unstack(): mengubah series menjadi tabel

# Mulai

# Kita mulai dengan mengimpor modul pandas dan data csv tersebut
import pandas as pd
df = pd.read_csv("C:/Users/safiq/Downloads/P05-01.csv")
```

```python
# 1. Histogram distribusi total nominal yang belum dibayar pengguna dengan status aktif (Existring Customer), dengan bins 20
df.loc[df["Attrition_Flag"] == "Existing Customer"]["Total_Revolving_Bal"].plot(kind="hist", bins=20, rwidth=0.8)
```

```python
# 2. Pie chart perbandingan banyaknya masing-masing jenjang pendidikan terakhir dari pengguna dengan banyak transaksi selama 12 bulan terakhir di atas 30
df.loc[df["Total_Trans_Ct"] > 30]["Education_Level"].value_counts().plot(kind="pie", autopct=lambda p:f'{p:.2f}%', ylabel="")
```

```python
# 3. Scatter plot dengan usia pengguna sebagai x dan nominal limit kartu kredit sebagai y dari pengguna berjenis kelamin laki-laki (M)
df.loc[df["Gender"] == "M"].plot(kind="scatter", x="Customer_Age", y="Credit_Limit")
```

```python
# 4. Horizontal bar chart menggambarkan jumlah pengguna dari setiap tipe kartu
df["Card_Category"].value_counts().plot(kind="barh")
```

```python
# 5. Stacked bar chart dari pengguna dengan status perkawinan "Married", "Single", dan "Divorced" dengan jenis kelamin sebagai sumbu x dan status perkawinan sebagai stacked y
df.loc[(df["Marital_Status"] == "Married") | (df["Marital_Status"] == "Single") | (df["Marital_Status"] == "Divorced")].groupby(["Gender", "Marital_Status"])["CLIENTNUM"].size().unstack().plot(kind="bar", stacked=True)
```
